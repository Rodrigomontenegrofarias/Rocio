# Modelo Matematico Beysens 2016.(Literatura)

## Formula matematica de Beysens para calcular 



## 
* RRN(Rendimiento del Rocio Natural)= 0.37
* H(Altura) = ?
* H(Altura al cuadrado) = ?
* Temperatura de derretimiento= td
* Temperatura del aire =  ta

```python
import math
import numpy as np
H = 0.9
H_2= H**2 
V0= 4.4
a = 17.27;
b = 237.7;
c= np.log(df["HR (%)"])
alpha = a*((df["T (C)"])/(b + df["T (C)"]))+ c
Td = (b * alpha)/(a - alpha);

exp20=np.exp((-((df["Viento (m/s)"])/V0)**20))

A = ( 0.37* ( 1 + (0.204323 * H) - (0.0238893 * H_2) - (18.0132 - (1.04963*H) + 0.21891 * H_2 ) * (10**(-3)*Td)) * (((Td + 273.15)/285)**4) * (1 - (df["Nubosidad-octa(%)"])/8)) + (0.06 *(Td-(df["T (C)"]))) * (1 + 100 * (1- exp20))
df["calculo"] = A
df["sum(dh/dt)calc(mm)"] = A
df['sum(dh/dt)exp(mm)']=	df['Flujo (L/da)']
df
```
